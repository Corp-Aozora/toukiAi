```python
from pdfminer.high_level import extract_text
import re

def extract_text_by_keyword(pdf_path, keyword):
    # PDFファイルからテキストを抽出
    text = extract_text(pdf_path)
    
    # テキストを行に分割
    lines = text.split('\n')
    
    # キーワードを含む行を見つけ、次の次の行のテキストを取得
    for i, line in enumerate(lines):
        if keyword in line:
            # キーワードが含まれる行の次の次の行を取得
            target_line_index = i +2
            if target_line_index < len(lines):
                return lines[target_line_index]
    return None  # キーワードが見つからなかった場合

# PDFファイルのパスを指定
pdf_path = r'C:\Users\PC_User\Desktop\司法書士の仕事用\00完了案件\20210120西澤勇二（売買による所有権移転、相続／原田いく・西澤寿春）\\松前郡松前町字江良９５７－１不動産登記（土地全部事項）2021012000257227.PDF'

# 特定のキーワードを指定
keyword = '地　番'


# キーワードが含まれる行の次の次の行のテキストを取得して表示
target_text = extract_text_by_keyword(pdf_path, keyword)

pattern = r'(\d+)番(\d*)'
match = re.search(pattern, target_text)
if match:
    # 一致する文字列が見つかった場合
    matched_text = match.group()
    print("見つかった文字列:", matched_text)
else:
    # 一致する文字列が見つからなかった場合
    print("指定されたパターンに一致する文字列はありません。")
```
